```python
import json
import pandas as pd
```

```python
"""
with open("ticker_data_17FEB2021.jsonl") as datafile:
    data = json.load(datafile)
df = pd.DataFrame(data)
"""
data_list = []
for line in open('ticker_data_17FEB2021.jsonl', 'r'):
    data_list.append(json.loads(line))
```

```python
data_list[0]
```

```python
df = pd.DataFrame()
for i in range(len(data_list)):
    if (i%1000==0):
        print(i)
    dictionary = data_list[i]['ohlc']
    dictionary['date'] = data_list[i]['last_trade_time'][:10]
    dictionary['time'] = data_list[i]['last_trade_time'][11:]
    dictionary['instrument token'] = data_list[i]['instrument_token']
    dictionary['volume'] = data_list[i]['volume']
    dictionary['tradable'] = data_list[i]['tradable']
    dictionary['mode'] = data_list[i]['mode']
    dictionary['last_price'] = data_list[i]['last_price']
    dictionary['last_quantity'] = data_list[i]['last_quantity']
    dictionary['average_price'] = data_list[i]['average_price']
    dictionary['buy_quantity'] = data_list[i]['buy_quantity']
    dictionary['sell_quantity'] = data_list[i]['sell_quantity']
    dictionary['change'] = data_list[i]['change']
    dictionary['oi'] = data_list[i]['oi']
    dictionary['oi_day_high'] = data_list[i]['oi_day_high']
    dictionary['oi_day_low'] = data_list[i]['oi_day_low']
    dictionary['timestamp'] = data_list[i]['timestamp']
    for j in range(5):
        dictionary['depth-buy-quantity'+str(j+1)] = data_list[i]['depth']['buy'][j]['quantity']
        dictionary['depth-buy-price'+str(j+1)] = data_list[i]['depth']['buy'][j]['price']
        dictionary['depth-buy-orders'+str(j+1)] = data_list[i]['depth']['buy'][j]['orders']
        dictionary['depth-sell-quantity'+str(j+1)] = data_list[i]['depth']['sell'][j]['quantity']
        dictionary['depth-sell-price'+str(j+1)] = data_list[i]['depth']['sell'][j]['price']
        dictionary['depth-sell-orders'+str(j+1)] = data_list[i]['depth']['sell'][j]['orders']
    df = df.append(dictionary,ignore_index = True)
df.head()
```

```python
df = df.rename(columns={'open': 'ohlc-open', 'high': 'ohlc-high', 'low': 'ohlc-low', 'close': 'ohlc-close'})
df = df.sort_values(['date','time']).reset_index(drop=True)
df.head()
```

```python
df.to_csv('ticker_data_17FEB2021.csv',index=False)
```
